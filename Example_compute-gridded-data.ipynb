```python
import h5py
import numpy as np

import astro_helper as ah
from GriddedData import GriddedDataset

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import gridspec
from matplotlib import cm
from matplotlib import rc
plt.style.use('bmh')
cmap = mpl.colormaps['viridis']  # Example: 'viridis' colormap
rc('font',**{'family':'DejaVu Sans','sans-serif':['Helvetica']})
```

```python
data = GriddedDataset('NGC300', max_grid_width=12., snapname='snap-DESPOTIC_500.hdf5', ptl_resolution=10.)
```

```python
# rotation curve
Rbin_centers, vcs = data.get_rotation_curve()
```

```python
plt.plot(np.array(Rbin_centers)/ah.kpc_to_cm, np.array(vcs)/ah.kms_to_cms)
plt.xlabel('R/kpc')
plt.ylabel(r'$v_{\rm circ}/{\rm km~s}^{-1}$')
plt.show()
```

```python
# galactic angular velocity
Omegaz_array = data.get_Omegaz_array()
```

```python
print(np.shape(Omegaz_array))
plt.imshow(Omegaz_array, cmap=cmap)
plt.colorbar()
plt.show()
```

```python
plt.hist(np.ravel(Omegaz_array), bins=30, histtype='step', lw=2, log=True)
plt.show()
```

```python
# epicyclic frequency
kappa_array = data.get_kappa_array(polyno=2, wndwlen=41)
```

```python
print(np.shape(kappa_array))
plt.imshow(kappa_array, cmap=cmap)
plt.colorbar()
plt.show()
```

```python
plt.hist(np.ravel(kappa_array), bins=30, histtype='step', lw=2, log=True)
plt.show()
```

```python
# interstellar medium weight
weight_array = data.get_weight_array()
```

```python
print(np.shape(weight_array))
plt.imshow(np.log10(weight_array/ah.kB_cgs), cmap=cmap)
plt.colorbar()
plt.show()
```
